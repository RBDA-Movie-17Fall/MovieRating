```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/Blair/Desktop/rmse.txt', sep=" ", header=None)
data.columns = ["Decision-Tree Size", "RMSE"]
data["Decision-Tree Size"] = data["Decision-Tree Size"].map(lambda x: str(x)[:-1])
```

```python
import matplotlib.pyplot as plt
 
Y = data["RMSE"]         
X = data["Decision-Tree Size"]
plt.plot(X, Y,linewidth=5)          
 
plt.title('RMSE',fontsize=18)    
plt.xlabel('Decision Tree Size',fontsize=14)
plt.ylabel('RMSE VALUE',fontsize=14)
 
plt.tick_params(axis='both',labelsize=14)     
plt.show()                 
```

```python
X1=X.astype('float64', raise_on_error = False)
```

```python

xnew = np.linspace(X1[0],X1[10],22)
cs = CubicSpline(X1, Y)
plt.title('RMSE',fontsize=18)    
plt.xlabel('Decision Tree Size',fontsize=14)
plt.ylabel('RMSE VALUE',fontsize=14)
plt.plot(xnew, cs(xnew))

plt.savefig('smooth_RMSE.png')
plt.show()
```
